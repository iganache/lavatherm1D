```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import LavaFlow as flow
```

```python
Tatm = 737
Tsurf = np.arange(1400, 800, -50)
qrad = flow.radiation_flux(Tsurf, Tatm, emis_surf = 0.95)
qf = flow.flynn_flux(Tsurf)
qs = flow.coupled_flux(Tsurf, Tatm, k=1, emis_surf = 0.95)
```

```python
plt.figure()
plt.plot(Tsurf, qrad)
plt.plot(Tsurf, qf)
plt.plot(Tsurf, qs)
plt.show()
```

```python
fig, ax = plt.subplots(1,1)
fig.set_size_inches(8, 8)

ax.plot(Tsurf, qrad/1e6, label = "Mueller et al., 2007")
ax.plot(Tsurf, qf/1e6, label = "Flynn et al., 2003")
ax.plot(Tsurf, qs/1e6, label = "this study")

ax.set_xlabel("Surface temperature (K)", fontsize = 20)
ax.set_ylabel("Heat flux (MW /m$^2$)", fontsize = 20)
plt.legend(fontsize=20)
ax.tick_params(axis="both", which="major", labelsize=20)
ax.xaxis.labelpad = 15
ax.yaxis.labelpad = 15
# plt.gca().ticklabel_format(axis='y', style='sci', scilimits=(0,0)) 
plt.savefig("heatflux_comp.png", bbox_inches='tight')
plt.show()
```
